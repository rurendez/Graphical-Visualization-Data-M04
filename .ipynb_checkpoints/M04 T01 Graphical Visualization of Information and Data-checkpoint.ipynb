## Exercise #1

### Downloading the dataset


Download the dataset "preu lloguer per trimestre i barri.csv" (rental price per quarter and neighborhood) which have been extracted from OpenDataBCN web. Then, summarize it statistically and graphically.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

# Read the file located at same folder as jupyter file
rent = pd.read_csv("preu lloguer per trimestre i barri.csv", sep=',', encoding='utf8', index_col=0)

# Display de first rows 
rent.head()
```

```python
# Show column names
rent.columns
```

```python
# Rename all column names to english
rent.rename(columns={'Trimestre': 'Quarter', 'Codi_Districte': '#District', 'Nom_Districte': 'District', 
                     'Codi_Barri': '#Neighborhood', 'Nom_Barri': 'Neighborhood', 'Lloguer_mitja': 'Average_rent', 
                     'Preu': 'Price'}, inplace=True)
```

```python
# There are two columns "Average_rent" and "Price" that are mixing 2 concepts: Euro/m2 month and Euro/month
rent.sample(20)
```

```python
# Show the sape of the dataframe
rent.shape
```

```python
# Rename index name to "Year"
rent.index.name = "Year"
```

```python
# Let's find the uniques values for each column, so we can understand better the structure of the dataset
# Barcelona is divided in 10 Districts, and total neighborhood of Barcelona are 73
# This dataset contains information of the average monthly rent price, and average monthly rent price per 
# square meter. This info is available for each quarter of year 2021
rent.nunique()
```

```python
# We have 10 districts, 73 neighboorhoods and 4 quarter of info "Euro/m2 month" and "Euro/month"
# So lets have the 2 two prices in separates series
# Fisrt, let's sort the original dataset
rent.sort_values(by=['#District','#Neighborhood','Quarter'], inplace = True)

# Now we create a filtered dataset with only the "Euro/month" average price
rent_A = rent[rent['Average_rent'] == "Lloguer mitjà mensual (Euros/mes)"]
rent_A.rename(columns={'Price':'Euro/month'}, inplace=True) # Lets rename correcty the column filtered

rent_A.head()
```

```python
# Now we create a filtered dataset with only the "Euro/m2" average price
rent_B = rent[rent['Average_rent'] == "Lloguer mitjà per superfície (Euros/m2 mes)"]
rent_B.rename(columns={'Price':'Euro/m2 month'}, inplace=True) # Lets rename correcty the column filtered

rent_B.head()
```

```python
# Once both dataframes are ordered, we can concatecante "Euro/m2 month" from rent_B to rent_A and create the
# data set rent_v2
extracted_col = rent_B['Euro/m2 month']
rent_v2 = pd.concat([rent_A, extracted_col], axis=1)
rent_v2.drop('Average_rent', axis=1, inplace=True)
rent_v2.head()
```

```python
# Get info of the data set
rent_v2.info()
```

```python
# Change datatype of Price to numeric
rent_v2["Euro/month"] = rent_v2["Euro/month"].apply(pd.to_numeric, errors='coerce')
rent_v2["Euro/m2 month"] = rent_v2["Euro/m2 month"].apply(pd.to_numeric, errors='coerce')
```

```python
# Get info of the new dataframe to check dtatype of new renamed columns
rent_v2.info()
```

```python
# Once converted the price data, we are going to check the null values, 22 for each ot the 2 new columns
print(rent_v2.isnull().sum())
```

```python
# In orderto keep all neighboorhoods, I'm not going to drop all nan valudes, but give them a mode value
rent_v2['Euro/month'] = rent_v2['Euro/month'].fillna(rent_v2['Euro/month'].mode()[0])
rent_v2['Euro/m2 month'] = rent_v2['Euro/m2 month'].fillna(rent_v2['Euro/m2 month'].mode()[0])
```

```python
# Let's check we do not have more nan values
print(rent_v2.isnull().sum())
```

```python
# Sstatistical summary of the dataframe
rent_v2.describe()
```

### Display data 

####  Display One categorical variable (District or Neighborhood)


```python
# Lets show a countplot of total neightborhoods of each disctric, shown by quarter. The size of each quarter is the same
# because we have the same data for each quarter and district
f, ax = plt.subplots(figsize=(8,4))
ax = sns.countplot(y=("District"), data=rent_v2, hue = "Quarter", palette="husl", dodge = True)
ax.set(xlabel='Total Count for each Quarter',
       ylabel='Disctic Name')
plt.grid()
plt.title("Number of Neighborhoods per District", fontsize=20)
plt.tight_layout()
plt.savefig("1-Number of Neighborhoods per District.png")
plt.show()
```

####  Display One numerical variable (Price)

```python
# Let's show the density plot of the Euro/month price for Barcelona 2021
# In this chart we can say that the average rent price is near to 750 Euro/month
f, ax = plt.subplots(figsize=(8,4))
rent_p1 = rent_v2['Euro/month']
sns.kdeplot(data=rent_p1)
plt.grid()
plt.title('Price distribution', fontsize=20)
plt.xlabel('Euro/month')
plt.tight_layout()
plt.savefig("2-Price distribution.png")
plt.show()
```

####  One numerical variable and one categorical (Disctric and Price)

```python
# Now, let's create a boxplot, so we can see the behaviour of the Euro/month price in each Distric
f, ax = plt.subplots(figsize=(8,4))
sns.boxplot( y=rent_v2["District"], x=rent_v2["Euro/month"])
plt.grid()
plt.title('Average Rent Price per District')
plt.xlabel('Euro/month')
plt.tight_layout()
plt.savefig("3a-Average Rent Price per Distric3.png")
plt.show()
```

```python
# Now, let's create a boxplot, so we can see the behaviour of the Euro/m2/month price in each Distric
# It is interesting to check how the realitve situation of each group of values difers from the Euro/month. This has to do
# with the value on m2. This value of m2, can be deduced and we will check it later.
f, ax = plt.subplots(figsize=(8,4))
sns.boxplot( y=rent_v2["District"], x=rent_v2["Euro/m2 month"])
plt.grid()
plt.title('Average Rent Price per m2 per District', fontsize=20)
plt.xlabel('Euro/m2/month')
plt.tight_layout()
plt.savefig("3b-Average Rent Price per Distric3.png")
plt.show()
```

####  Two numerical variables (Year or Quarter and Price)

```python
# By means of a scatterplot we can check where are located the Euro/month price depending on the Quarter, with is ve similar
sns.set_theme(style="whitegrid")
f, ax = plt.subplots(figsize=(14,4)) #set the plot size
sns.scatterplot(x="Euro/month", y="Quarter",palette="ch:r=-.2,d=.3_r", data=rent_v2, ax=ax)
plt.title('Price ditribution during each quarter', fontsize=20)
plt.xlabel('Euro/month')
plt.tight_layout()
plt.savefig("4-Price ditribution during each quarter.png")
plt.show()
```

####  Three variables (Neighborhood or District, Trimestre and Price)

```python
# Let's extract from dataset the columns we want to work with
# We want to get a heatmap indicating de Euro/month mean for every District and Quarter.
g = rent_v2[["District","Quarter","Euro/month"]]
# Now we get the desire grup and obtain the mean ot hte researched variable
h = g.groupby(["District", "Quarter"], as_index = False)[["Euro/month"]].mean()
h.head()
```

```python
# Let's pivot the table in order to prepare the data for the heatmap graphic
newh = h.pivot(index='District', columns='Quarter')
newh.head()
```

```python
# Now let's show our heatmap
# Let's set the palette to show the lower prices in yelow and the highest in red. As we can see the darker red are the 
# neighborhoods more expensive, and the softest the less expensive. 
sns.color_palette("YlOrRd", as_cmap=True)
# Set the space Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(newh, annot=True, fmt="g", cmap="YlOrRd") # Set color palette
plt.title('Average Price per Disctrict per Quarter 2021', fontsize=20)
plt.xlabel('Quarter')
plt.tight_layout()
plt.savefig("5-Average Price per Disctrict per Quarter 2021.png")
plt.show()
```

## Exercise #3
Proposing any other visualization that may be interesting.

```python
# From the dataset we have 2 types of data, Euro(months and Euro/m2 month. All averages. So we can get the m2 average. Then,
# we will be able to show a new heatmap showing that value in relation with all Neighbohoods.
rent_v2["m2"] = (rent_v2["Euro/month"] / rent_v2["Euro/m2 month"])
rent_v2.head()
```

```python
# Let's extract from dataset the columns we want to work with
# We want to get a heatmap indicating de m2 mean for every District and Quarter.
g = rent_v2[["Neighborhood","Quarter","m2"]]
# Now we get the desire grup and obtain the mean ot hte researched variable
h = g.groupby(["Neighborhood", "Quarter"], as_index = False)[["m2"]].mean()
```

```python
# Let's pivot the table in order to prepare the data for the heatmap graphic
newh = h.pivot(index='Neighborhood', columns='Quarter')
newh.head()
```

```python
# Now let's show our heatmap
# Let's set the palette to show the lower prices in yelow and the highest in red

# Set the space Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(14, 22))
sns.heatmap(newh, annot=True, fmt="g", cmap="YlOrRd") # Set color palette
plt.title('Average m2 Price per Neighborhood per Quarter 2021', fontsize=20)
plt.xlabel('Quarter')
plt.tight_layout()
plt.savefig("6-Average m2 Price per Neighborhood per Quarter 2021.png")
plt.show()
```
